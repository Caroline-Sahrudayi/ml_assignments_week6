```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("run_or_walk.csv")
df.head(5)
```

```python
df.shape
```

```python
print('Missing values:\n{}'.format(df.isnull().sum()))

# Find duplicated records
print('\nNumber of duplicated records: {}'.format(df.duplicated().sum()))
```

```python
df.describe()
```

```python
df.info()
```

```python
df=df.drop(['username','date','time'],axis=1)
```

```python
y=df['activity']
x=df.drop(['activity'],axis=1)
```

```python
x.head(5)
```

```python
from sklearn.naive_bayes import GaussianNB
import seaborn as sns # this is used for the plot the graph
from sklearn.model_selection import train_test_split # to split the data into two parts
from sklearn import metrics # for the check the error and accuracy of the model
from sklearn import tree
from sklearn.metrics import confusion_matrix
 
```

```python
x_train, x_test, y_trx_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state = 42)
```

```python
gnb = GaussianNB()
#mnb = MultinomialNB()
y_pred_gnb = gnb.fit(x_train,y_train).predict(x_test)
#target_pred = y_pred_mnb.predict(test_X)
print(metrics.accuracy_score(y_test, y_pred_gnb))
cnf_matrix_gnb = confusion_matrix(y_test, y_pred_gnb)
print(cnf_matrix_gnb)
print(metrics.accuracy_score(y_pred_gnb,y_test))
```

```python
print(metrics.classification_report(y_test,y_pred_gnb))
```
